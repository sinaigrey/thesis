```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the .h5ad file
adata = sc.read_h5ad("all_cells.h5ad")

# Function to plot and test correlation for one treatment
def plot_gene_correlation_one(gene1, gene2, adata_subset, treatment):
    data1 = adata_subset[adata_subset.obs['treatment'] == treatment][:, gene1].X
    data2 = adata_subset[adata_subset.obs['treatment'] == treatment][:, gene2].X

    correlation, p_value = pearsonr(data1.flatten(), data2.flatten())

    plt.scatter(data1, data2, s=10)
    plt.title(f'{treatment}:\nCorrelation between {gene1} and {gene2}: {correlation:.2f}, p-value: {p_value:.2e}')
    plt.xlabel(gene1)
    plt.ylabel(gene2)
    plt.show()

# Function to plot and test correlation for two treatments
def plot_gene_correlation_two(gene1, gene2, adata_subset, treatment1, treatment2):
    data1_t1 = adata_subset[adata_subset.obs['treatment'] == treatment1][:, gene1].X
    data2_t1 = adata_subset[adata_subset.obs['treatment'] == treatment1][:, gene2].X
    data1_t2 = adata_subset[adata_subset.obs['treatment'] == treatment2][:, gene1].X
    data2_t2 = adata_subset[adata_subset.obs['treatment'] == treatment2][:, gene2].X

    correlation_t1, p_value_t1 = pearsonr(data1_t1.flatten(), data2_t1.flatten())
    correlation_t2, p_value_t2 = pearsonr(data1_t2.flatten(), data2_t2.flatten())

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    axs[0].scatter(data1_t1, data2_t1, s=1)
    axs[0].set_title(f'{treatment1}:\nCorrelation: {correlation_t1:.2f}, p-value: {p_value_t1:.2e}')
    axs[0].set_xlabel(gene1)
    axs[0].set_ylabel(gene2)

    axs[1].scatter(data1_t2, data2_t2, s=1)
    axs[1].set_title(f'{treatment2}:\nCorrelation: {correlation_t2:.2f}, p-value: {p_value_t2:.2e}')
    axs[1].set_xlabel(gene1)
    axs[1].set_ylabel(gene2)

    plt.suptitle(f'Correlation of {gene1} vs {gene2} in {cell_type}')
    plt.show()

# Get user input for genes
gene1 = input("Enter the first gene: ")
gene2 = input("Enter the second gene: ")

# Get user input for cell type
cell_type = input("Enter the cell type or 'all' for a general comparison: ")

if cell_type.lower() == 'all':
    adata_subset = adata
else:
    adata_subset = adata[adata.obs['General Celltypes'] == cell_type]

# Get user input for comparison type
comparison_type = input("Would you like to compare within one treatment or between two treatments? (one/two): ").lower()

if comparison_type == 'one':
    # Get user input for treatment group
    treatment = input("Enter the treatment type: ")
    plot_gene_correlation_one(gene1, gene2, adata_subset, treatment)
elif comparison_type == 'two':
    # Get user input for treatment groups
    treatment1 = input("Enter the first treatment type: ")
    treatment2 = input("Enter the second treatment type: ")
    plot_gene_correlation_two(gene1, gene2, adata_subset, treatment1, treatment2)
else:
    print("Invalid comparison type selected.")
```

```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the .h5ad file
adata = sc.read_h5ad("all_cells.h5ad")

# Function to plot and test correlation for three treatments
def plot_gene_correlation_three(gene1, gene2, adata_subset, treatment1, treatment2, treatment3):
    data1_t1 = adata_subset[adata_subset.obs['treatment'] == treatment1][:, gene1].X
    data2_t1 = adata_subset[adata_subset.obs['treatment'] == treatment1][:, gene2].X
    data1_t2 = adata_subset[adata_subset.obs['treatment'] == treatment2][:, gene1].X
    data2_t2 = adata_subset[adata_subset.obs['treatment'] == treatment2][:, gene2].X
    data1_t3 = adata_subset[adata_subset.obs['treatment'] == treatment3][:, gene1].X
    data2_t3 = adata_subset[adata_subset.obs['treatment'] == treatment3][:, gene2].X

    correlation_t1, p_value_t1 = pearsonr(data1_t1.flatten(), data2_t1.flatten())
    correlation_t2, p_value_t2 = pearsonr(data1_t2.flatten(), data2_t2.flatten())
    correlation_t3, p_value_t3 = pearsonr(data1_t3.flatten(), data2_t3.flatten())

    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    
    sc1 = axs[0].scatter(data1_t1, data2_t1, s=10, c=data1_t1, cmap='coolwarm')
    axs[0].set_title(f'{treatment1}:\nCorrelation: {correlation_t1:.2f}, p-value: {p_value_t1:.2e}')
    axs[0].set_xlabel(gene1)
    axs[0].set_ylabel(gene2)
    plt.colorbar(sc1, ax=axs[0], orientation='horizontal')

    sc2 = axs[1].scatter(data1_t2, data2_t2, s=10, c=data1_t2, cmap='coolwarm')
    axs[1].set_title(f'{treatment2}:\nCorrelation: {correlation_t2:.2f}, p-value: {p_value_t2:.2e}')
    axs[1].set_xlabel(gene1)
    axs[1].set_ylabel(gene2)
    plt.colorbar(sc2, ax=axs[1], orientation='horizontal')

    sc3 = axs[2].scatter(data1_t3, data2_t3, s=10, c=data1_t3, cmap='coolwarm')
    axs[2].set_title(f'{treatment3}:\nCorrelation: {correlation_t3:.2f}, p-value: {p_value_t3:.2e}')
    axs[2].set_xlabel(gene1)
    axs[2].set_ylabel(gene2)
    plt.colorbar(sc3, ax=axs[2], orientation='horizontal')

    plt.suptitle(f'Correlation of {gene1} vs {gene2} in {cell_type}')
    plt.show()

# Get user input for genes
gene1 = input("Enter the first gene: ")
gene2 = input("Enter the second gene: ")

# Get user input for cell type
cell_type = input("Enter the cell type or 'all' for a general comparison: ")

if cell_type.lower() == 'all':
    adata_subset = adata
else:
    adata_subset = adata[adata.obs['General Celltypes'] == cell_type]

# Get user input for subcluster
subcluster = input("Enter the subcluster or 'none' if you don't want a subcluster: ")

if subcluster.lower() != 'none':
    adata_subset = adata_subset[adata_subset.obs['cluster'] == subcluster]

# Get user input for comparison type
comparison_type = input("Would you like to compare within one treatment, between two treatments, or three treatments? (one/two/three): ").lower()

if comparison_type == 'one':
    # Get user input for treatment group
    treatment = input("Enter the treatment type: ")
    plot_gene_correlation_one(gene1, gene2, adata_subset, treatment)
elif comparison_type == 'two':
    # Get user input for treatment groups
    treatment1 = input("Enter the first treatment type: ")
    treatment2 = input("Enter the second treatment type: ")
    plot_gene_correlation_two(gene1, gene2, adata_subset, treatment1, treatment2)
elif comparison_type == 'three':
    # Get user input for treatment groups
    treatment1 = input("Enter the first treatment type: ")
    treatment2 = input("Enter the second treatment type: ")
    treatment3 = input("Enter the third treatment type: ")
    plot_gene_correlation_three(gene1, gene2, adata_subset, treatment1, treatment2, treatment3)
else:
    print("Invalid comparison type selected.")
```

```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the .h5ad file in chunks to handle memory constraints
adata = sc.read_h5ad("all_cells.h5ad")

def plot_gene_correlation_chunked(gene1, gene2, adata_subset, treatments):
    fig, axs = plt.subplots(1, len(treatments), figsize=(18, 5))
    
    for idx, treatment in enumerate(treatments):
        data1 = adata_subset[adata_subset.obs['treatment'] == treatment][:, gene1].X
        data2 = adata_subset[adata_subset.obs['treatment'] == treatment][:, gene2].X
        correlation, p_value = pearsonr(data1.flatten(), data2.flatten())
        
        sc_plot = axs[idx].scatter(data1, data2, s=5, c=data1, cmap='coolwarm')
        axs[idx].set_title(f'{treatment}:\nCorrelation: {correlation:.2f}, p-value: {p_value:.2e}')
        axs[idx].set_xlabel(gene1)
        axs[idx].set_ylabel(gene2)
        plt.colorbar(sc_plot, ax=axs[idx], orientation='horizontal')

    plt.suptitle(f'Correlation of {gene1} vs {gene2} in {cell_type}')
    plt.show()

# Get user input
gene1 = input("Enter the first gene: ")
gene2 = input("Enter the second gene: ")
cell_type = input("Enter the cell type or 'all' for a general comparison: ")

if cell_type.lower() == 'all':
    adata_subset = adata
else:
    adata_subset = adata[adata.obs['General Celltypes'] == cell_type]

subcluster = input("Enter the subcluster or 'none' if you don't want a subcluster: ")

if subcluster.lower() != 'none':
    adata_subset = adata_subset[adata_subset.obs['cluster'] == subcluster]

comparison_type = input("Would you like to compare within one treatment, between two treatments, or three treatments? (one/two/three): ").lower()

if comparison_type == 'one':
    treatment = input("Enter the treatment type: ")
    plot_gene_correlation_chunked(gene1, gene2, adata_subset, [treatment])
elif comparison_type == 'two':
    treatment1 = input("Enter the first treatment type: ")
    treatment2 = input("Enter the second treatment type: ")
    plot_gene_correlation_chunked(gene1, gene2, adata_subset, [treatment1, treatment2])
elif comparison_type == 'three':
    treatment1 = input("Enter the first treatment type: ")
    treatment2 = input("Enter the second treatment type: ")
    treatment3 = input("Enter the third treatment type: ")
    plot_gene_correlation_chunked(gene1, gene2, adata_subset, [treatment1, treatment2, treatment3])
else:
    print("Invalid comparison type selected.")
```
